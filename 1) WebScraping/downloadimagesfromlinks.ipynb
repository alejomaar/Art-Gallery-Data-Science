# Dowload images from links

```python
import pandas as pd
import requests

df = pd.read_excel("googleArtFinal.xlsx")
#df = df.head(10)
imageLinks = df.weblink
filename = []

indexImage=0
for link in imageLinks:
    if(link.startswith('https://')):
        currentFilename = "google_images\img"+str(indexImage)+".jpg"
        filename.append("img"+str(indexImage)+".jpg") 
        img_data = requests.get(link).content
        with open(currentFilename, 'wb') as handler:
            handler.write(img_data)
    else:
        filename.append("null") 
    indexImage=indexImage+1
    
df["filename"] = filename
df.to_csv (r'googlefullcomplete4k.csv', index = False)

```

```python
%matplotlib inline
#The line above is necesary to show Matplotlib's plots inside a Jupyter Notebook

import cv2
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import os
import glob
import os.path
import re

categories = ["WHITE","PINK","YELLOW","PURPLE","BLUE","TEAL","GREEN","ORANGE","RED","BROWN","BLACK"]

data = {"name":[],"class":[],"MeanR":[],"MeanG":[],"MeanB":[],"StdR":[],"StdG":[],"StdB":[]}
for imgroute in os.listdir('google_images'):
    currentCategory = None
    for category in categories:
        if category in imgroute:
            currentCategory = category
    data["name"].append(imgroute)
    data["class"].append(currentCategory)
    
    route = "C:/Users/user/BigData/google_images/"+imgroute
    print(imgroute)
    if(os.path.isfile(route)):
        image = cv2.imread(route)
        (B, G, R) = cv2.split(image)
        data["MeanR"].append(np.mean(R))
        data["MeanG"].append(np.mean(G))
        data["MeanB"].append(np.mean(B))
        data["StdR"].append(np.std(R))
        data["StdG"].append(np.std(G))
        data["StdB"].append(np.std(B))
    
print(data)



```
