### Funciones

```python
def df_style(samples):
    #Crear df
    df_n = pd.DataFrame(columns=['#imag' ,'mean','R_mean','G_mean', 'B_mean', 'std', 'R_std','G_std', 'B_std', 'asi', 'R_asi', 'G_asi', 'B_asi'])
    xn=0
    for x in samples:
        if xn != 3827: #no existia la imagen 38.
            img_ = "".join(x)
            ruta = ("google_images/"+img_)
            img_ = cv2.imread(ruta)
            mEstad = DatosEstColor(img_)
            df_n = df_n.append({'#imag':x , 'mean':mEstad[0],'R_mean':mEstad[1],'G_mean':mEstad[2], 'B_mean':mEstad[3], 'std':mEstad[4], 'R_std':mEstad[5],'G_std':mEstad[6], 'B_std':mEstad[7],  'asi':mEstad[8], 'R_asi':mEstad[9],'G_asi':mEstad[10], 'B_asi':mEstad[11]}  ,ignore_index=True)
        xn=xn+1
    return df_n
```

```python
from scipy.stats import skew
def DatosEstColor(img):
    (B, G, R) = cv2.split(img) #separa la imagen en us 3 canales
    m = img.mean()    #media
    std_ = img.std()  #desv estandar
    asi = float(skew(cv2.calcHist([img],[0],None,[256],[0,256]))) #asimetria
    
    mR = R.mean()
    stdR = R.std()
    asiR = float(skew(cv2.calcHist([R],[0],None,[256],[0,256]))) #asimetria
    
    mG = G.mean()
    stdG = G.std()
    asiG = float(skew(cv2.calcHist([G],[0],None,[256],[0,256]))) #asimetria
    
    mB = B.mean()
    stdB = B.std()
    asiB = float(skew(cv2.calcHist([B],[0],None,[256],[0,256]))) #asimetria
    
    mEst = [m, mR, mG, mB, std_, stdR, stdG, stdB, asi, asiR, asiG, asiB]

    return mEst #Retorna vector de caracteristicas
```
